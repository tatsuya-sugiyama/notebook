```
def count_param(model : nn.Module):
  param = model.named_parameters()
  itr = [np.prod(v.size()) for name, v in param if "auxiliary" not in name]
  return np.sum(itr)
```

```
model = nn.Conv2d(3, 64, 3)
count_param(model)
```
